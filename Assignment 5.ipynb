## Q4

```python
x=np.linspace(-10,10,100)
```

```python
y1=x**2
plt.figure(figsize=(6,4))
plt.grid(True)
plt.plot(x, y1, label="Quadratic")
plt.title('Quadratic Funtion')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
y2=np.sin(x)
plt.figure(figsize=(10,2))
plt.grid(True)
plt.plot(x, y2, label="Sine Funtion")
plt.title('Sine Funtion')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
y3=np.exp(x)
plt.figure(figsize=(6,4))
plt.grid(True)
plt.plot(x, y3, label="Exponential Funtion")
plt.title('Exponential Funtion')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
y4=np.log(abs(x)+1)
plt.figure(figsize=(6,4))
plt.grid(True)
plt.plot(x, y4, label="logarithmic Funtion")
plt.title('Logarithmic Funtion')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
