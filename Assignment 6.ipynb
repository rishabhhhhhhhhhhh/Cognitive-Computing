```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1

```python
x=np.linspace(-10,10,101)
```

```python
M=int(input("Enter a number:"))
```

```python
y1=M*x**2
y2=M*np.sin(x)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y1,'b--',label="M*X^2")
plt.plot(x,y2,'r-',label="M*sin(x)")
plt.legend()
plt.grid()
plt.title('Graphs')
plt.show()
```

## Q2

```python
#dataset of 5 subjects
subjects = ["Math", "OS", "DBMS", "DSA", "AI"]
scores = [85, 78, 92, 88, 75]
df=pd.DataFrame({'Subjects':subjects,'Scores':scores})
df
```

```python
plt.figure(figsize=(6,5))
barplot=sns.barplot(x="Subjects",y="Scores",data=df,palette="mako")
for index, value in enumerate(scores):
    plt.text(index, value+1,str(value),ha='center',fontsize=12,fontweight='bold')
plt.title("Scores in Different Subjects")
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y',linestyle="--",alpha=0.5)
plt.show()
```

## Q3

```python
np.random.seed(102317044)
data=np.random.randn(50)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))
cumulative_sum = np.cumsum(data)
axs[0].plot(cumulative_sum, color='blue')
axs[0].set_title('Cumulative Sum Line Plot')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)
noise = np.random.normal(0, 1, 50)
axs[1].scatter(data, noise, color='red')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Data Points')
axs[1].set_ylabel('Noise')
axs[1].grid(True)
plt.show()
```

```python
sales=pd.read_csv('company_sales_data.csv')
sales
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=sales, marker='o', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```
