```python
def rate(x):
    if x>=4.5:
        return 'Excellent'
    elif x>=4 and x<4.5:
        return 'Good'
    else:
        return 'Average'
empdata['Performance']=empdata['Rating'].apply(rate)
empdata
```

```python
def tax(x):
    deduct=0.1*x
    return deduct
empdata['Tax']=empdata['Salary'].apply(tax)
empdata
```
